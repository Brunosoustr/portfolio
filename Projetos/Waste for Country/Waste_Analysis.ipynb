# Análise de Desperdício de Alimentos - Waste for Country

Este notebook contém a análise de dados do projeto Waste for Country, que investiga o desperdício de alimentos em diferentes países.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset
df = pd.read_excel('Waste.xlsx')

# Exibir as primeiras linhas do dataframe
df.head()
```

## 1. Análise Geral do Desperdício

### Qual é a categoria alimentar que gera o maior desperdício total em toneladas e o maior prejuízo econômico?

```python
category_waste = df.groupby('Food Category')['Total Waste (Tons)'].sum().sort_values(ascending=False)
category_loss = df.groupby('Food Category')['Economic Loss (Million $)'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=category_waste.index, y=category_waste.values)
plt.title('Total de Desperdício por Categoria de Alimento')
plt.xlabel('Categoria de Alimento')
plt.ylabel('Total de Desperdício (Tons)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=category_loss.index, y=category_loss.values)
plt.title('Prejuízo Econômico por Categoria de Alimento')
plt.xlabel('Categoria de Alimento')
plt.ylabel('Prejuízo Econômico (Milhões $)')
plt.xticks(rotation=45)
plt.show()
```

### Existe uma relação direta entre o total de resíduos gerados e a perda econômica?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Total Waste (Tons)', y='Economic Loss (Million $)', hue='Food Category')
plt.title('Relação entre Desperdício Total e Perda Econômica')
plt.xlabel('Total de Desperdício (Tons)')
plt.ylabel('Perda Econômica (Milhões $)')
plt.show()
```
